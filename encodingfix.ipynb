```python
# zouska na testovacich datech
walk = os.walk("data_test/dbf", followlinks=0)
files = []


def convertdbftocsv(file):
    codecs = ("cp852","cp437","utf-8","cp1250")
    for codec in codecs:
        try:
            dbf = simpledbf.Dbf5(file, codec=codec)
            csv = file + "." + codec + ".csv"
            dbf.to_csv(csv)
            #print(file, "coverted to", csv)
            if codec != 'utf-8':
                fixencoding(csv,codec)
        except BaseException as err:
            print("dbftocsv",file, err, "with", codec)



def fixencoding(file,code):
    text = ""
    backup = file + "." + code + ".bak"
    try:
        with open(file, 'r', encoding=code) as f:
            text = f.read()
        try:
            shutil.copy(file,backup)
            print("writing backup file:", backup)
        except BaseException as err:
            print("error writing backup file", err, backup)

        with open(file, 'w', encoding='utf8') as fb:
            fb.write(text)
            print("\t",file, "converting from", code, "into utf-8:", file)
    except BaseException as err:
        print("\tfixencoding:", file, err)

def main():
    for a, b, c in walk:
        for j in c:
            # print(j)
            f = os.path.join(a,j)
            files.append(f)
    for datafile in files:
        _, filetype = datafile.rsplit(".", maxsplit=1)
        filetype = filetype.lower()
        if ("csv" or "xml" or "xsd") in filetype:
            try:
                with open(datafile,'r') as f:
                    f.read()
            except UnicodeDecodeError as unierr:
                print("main:", file, unierr)
                fixencoding(datafile)
        elif "dbf" in filetype:
            print(datafile)
            convertdbftocsv(datafile)
        # match filetype:
        #     case "bak":
        #         continue
        #     case "dbf":
        #         convertdbftocsv(datafile)
        #     case "csv" | "xml" | "xsd":
        #         try:
        #             with open(datafile,'r') as f:
        #                 text = f.read()
        #                 #print(jsonText)
        #         except UnicodeDecodeError:
        #             fixencoding(datafile)


main()
```
